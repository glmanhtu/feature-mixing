```python
from glob import glob
from PIL import Image
import numpy as np
```

```python
orig = '/data/mpeer/papyri_1200/croppedImages/'
binarized_imgs = sorted(list(glob('/data/mpeer/papyri_1200/binarized_unet_2/**/*.jpg', recursive=True)))
print(binarized_imgs)
```

```python
def get_orig(imgp):
    return imgp.replace('binarized_unet_2', 'croppedImages')
def get_sauvola(imgp):
    return imgp.replace('binarized_unet_2', 'binarized_sauvola')

def get_tuple(imgp):
    img_bin = Image.open(imgp)
    img_orig =Image.open(get_orig(imgp)).resize(img_bin.size)
    img_sauv =Image.open(get_sauvola(imgp)).resize(img_bin.size)

    return np.hstack([np.asarray(img_orig), np.asarray(img_sauv), np.asarray(img_bin)])

def get_row(idx, typ='orig'):
    imgs = []
    imgs.append(Image.open(get_orig(binarized_imgs[idx])).resize((512,128)))
    imgs.append(Image.open(get_sauvola(binarized_imgs[idx])).resize((512,128)))
    imgs.append(Image.open(binarized_imgs[idx]).resize((512,128)))
    return np.vstack([np.asarray(img) for img in imgs])
```

```python
idxs = [1,10,100,1000,1200,2000,2500,3000,4000,5000,4500,2800,900,3500-1]
typ='orig'
for idx in idxs:
    print(idx)
    img = Image.fromarray(get_row(idx))
    img.save(f'{idx}.png')
    # import matplotlib.pyplot as plt
    # plt.imshow(img)
    # plt.show()
    # break
    # save(f'binarized/{idx}.png')
```
